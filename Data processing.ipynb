```python
df_gameplaydata['event_day']=pd.to_datetime(df_gameplaydata['event_time'], format='%Y-%m-%d').dt.day
df_gameplaydata['event_month']=pd.to_datetime(df_gameplaydata['event_time'], format='%Y-%m-%d').dt.month
df_gameplaydata['event_year']=pd.to_datetime(df_gameplaydata['event_time'], format='%Y-%m-%d').dt.year
df_transactiondata['purchase_day']=pd.to_datetime(df_transactiondata['purchase_date'], format='%Y-%m-%d').dt.day
df_transactiondata['purchase_month']=pd.to_datetime(df_transactiondata['purchase_date'], format='%Y-%m-%d').dt.month
df_transactiondata['purchase_year']=pd.to_datetime(df_transactiondata['purchase_date'], format='%Y-%m-%d').dt.year
df_userdata['signup_day']=pd.to_datetime(df_userdata['signup_date'], format='%Y-%m-%d').dt.day
df_userdata['signup_month']=pd.to_datetime(df_userdata['signup_date'], format='%Y-%m-%d').dt.month
df_userdata['signup_year']=pd.to_datetime(df_userdata['signup_date'], format='%Y-%m-%d').dt.year
```

```python
#Nguoi choi chi tieu tien nhieu nhat la ai va level_completed ho dat duoc la bao nhieu
df1 = df_dtanalyst.groupby('user_id')['purchase_amount'].sum().sort_values(ascending=False)
df2 = df_dtanalyst.groupby('user_id')['level_completed'].max().sort_values(ascending=False)
df3 = pd.merge(df1, df2, on='user_id')
df4 = pd.merge(df3, df_userdata, on='user_id')
df4 = df4[['user_id', 'purchase_amount', 'level_completed', 'country', 'device']]
df4.to_csv('/content/drive/MyDrive/Project Data Analyst/data.csv')
df4.info()
df4.head(10)
```

```python
#Ty le phan tram nguoi choi con choi sau 30 ngay
df5 = df_dtanalyst.groupby('user_id')['event_time'].max()
df5 = pd.DataFrame(df5)
df5.columns = ['event_time_last']
df5 = pd.merge(df5, df_userdata, on='user_id')
df5['Number_of_days_played'] = (pd.to_datetime(df5['event_time_last']) - pd.to_datetime(df5['signup_date'])).dt.days
df5.to_csv('/content/drive/MyDrive/Project Data Analyst/data_tylengchoi.csv')
df5 = df5[df5['Number_of_days_played'] > 30]
df5.info()
tlccs30n = df5['user_id'].count() / df_userdata['user_id'].count() * 100
print(f'Ty le % nguoi con choi sau 30 ngay la: {tlccs30n}%')
```

```python
#Du doan doanh thu moi nguoi choi dua tren purchase_amount, level_completed, country, device su dung mo hinh LinearRegression
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
df4.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
df4['country'] = lb_make.fit_transform(df4['country'])
df4['device'] = lb_make.fit_transform(df4['device'])
df4.head()
```

```python
features = ['country', 'device', 'level_completed']
target = ['purchase_amount']
X = df4[features]
y = df4[target]
#Chia bo du lieu thanh hai tap train, test 90% train, 10%test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
#Huan luyen mo hinh
model = LinearRegression()
model.fit(X_train, y_train)
#Du doan doanh thu
y_pred = model.predict(X_test)
#Danh gia mo hinh
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```
